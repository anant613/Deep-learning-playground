```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Making the dataset
Data = {
    'Customer': ['Mukesh', 'Akshay', 'Raj', 'Dabloo', 'Rounit', 'Rohan'],
    'Age': [25, 30, 22, 35, 28, 40],
    'Annual Income': [50000, 60000, 45000, 70000, 52000, 80000]
}

df = pd.DataFrame(Data)

# Selecting features for clustering
X = df[['Age', 'Annual Income']]

# Using KMeans
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
df['Group'] = kmeans.fit_predict(X)

# Plotting the graph (basic version)
plt.figure(figsize=(6,5))
for group in df['Group'].unique():
    group_data = df[df['Group'] == group]
    plt.scatter(group_data['Age'], group_data['Annual Income'], label=f'Group {group}')

plt.title('Customer Segmentation using KMeans Clustering')
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.legend()
plt.show()

```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Customer': ['Mukesh', 'Akshay', 'Raj', 'Dabloo', 'Rounit', 'Rohan'],
    'Age': [25, 30, 22, 35, 28, 40],
    'Annual Income': [50000, 60000, 45000, 70000, 52000, 80000],
    'Spending Score': [50, 60, 40, 70, 55, 80]
}

df = pd.DataFrame(data)

# Data Preprocessing
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['Age', 'Annual Income', 'Spending Score']])

# Principal Component Analysis
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(pca_result, columns=['PCA1', 'PCA2'])
explained_variance = pca.explained_variance_ratio_

print("Explained variance by each component:")
print(np.round(explained_variance * 100, 2))

# Plotting the graph (basic)
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], c='blue', alpha=0.5)
plt.title('PCA of Customer Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid()
plt.show()

print("New Data with 2 features PCA1 and PCA2:")
print(pca_df)
```
