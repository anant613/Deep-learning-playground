### Indexing (selecting data from tensors)
indexing with pytorch is similar to indexing with numpy 


```python
#Create a tensor 

import torch 

x = torch.arange(1, 10).reshape(1,3,3)
x , x.shape 
```

```python
#lets index our first brarckt
x[0]
x[0][2]
x[0][2][2]
# you can use ":" to select "all" of a target dimension
x[:, 0]

# Index on x to return 9
print(x[0][2][2])

# index on x to return 3,6,9
print(x[:, 2])
```

### Trying to take randomm out of random 

In short how a neural network learns :

`start with random numbers -> tensor -> update random numbers to try and make them better representation of data -> againn -> again -> again...

To reduce the randomness in neural and Pytorch comes the concept of a **random seed** 

Essentially ehst the random seed does is a "flavour" the randomness

```python
import torch 

#Create two random tesnsor 
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3, 4)
print(random_tensor_A)
print(random_tensor_B)

print(random_tensor_A == random_tensor_B)
```

```python
# Lets make some random but reproducible tensors 

import torch

# set the problem seed 

RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)

random_tensor_C = torch.rand(3,4)
torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3,4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
